```python
import requests                     #Телеграм БОТ
from telegram.ext import Updater, CommandHandler, MessageHandler, Filters

# Команда-обработчик /start
def start(update, context):
    update.message.reply_text('Привет! Я чат-бот от B4DoS епт. Как сам Вывозишь? \n Мои команды: \n /help-Помощь \n /quote-Цитата \n /weather-Погода')
                             

# Команда-обработчик /help
def help(update, context):
    update.message.reply_text('Если у вас возникли вопросы, просто напишите мне ну или Набери на номер сам знаешь какой Епт')

# Ответ на текстовое сообщение
def reply_to_message(update, context):
    update.message.reply_text('Принято. Вы написали: ' + update.message.text)

# Команда-обработчик /quote
def quote(update, context):
    response = requests.get('https://api.quotable.io/random')
    if response.status_code == 200:
        data = response.json()
        quote = data['content'] + " - " + data['author']
        update.message.reply_text(quote)
    else:
        update.message.reply_text('Извините, не удалось получить цитату.')

# Команда-обработчик /weather
def weather(update, context):
    api_key = 'f6f297b530574a1da76113736232411'
    city = 'Simferopol'
    url = f'http://api.weatherapi.com/v1/current.json?key={api_key}&q={city}'
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        weather_description = data['current']['condition']['text']
        temperature = data['current']['temp_c']
        humidity = data['current']['humidity']
        wind_speed = data['current']['wind_kph']
        weather_info = f"Симферополь: {weather_description}, Температура:{temperature}°C, Влажность: {humidity}%, Скорость ветра: {wind_speed} км/ч"
        update.message.reply_text(weather_info)
    else:
        update.message.reply_text('Извините, не удалось получить информацию о погоде.')

def main():
    updater = Updater("6368420018:AAEA5twfklDSaxspo7XsO6DeSVhShgBKyBg", use_context=True)

    dp = updater.dispatcher
    dp.add_handler(CommandHandler("start", start))
    dp.add_handler(CommandHandler("help", help))
    dp.add_handler(CommandHandler("quote", quote))
    dp.add_handler(CommandHandler("weather", weather))
    dp.add_handler(MessageHandler(Filters.text & ~Filters.command, reply_to_message))

    updater.start_polling()
    updater.idle()

if __name__ == '__main__':
    main()
```
