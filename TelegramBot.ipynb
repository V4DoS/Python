```python
import telegram
from telegram import Update, ChatAction
from telegram.ext import Updater, MessageHandler, Filters, CallbackContext
import g4f


g4f.debug.logging = True
print(g4f.Provider.Ails.params)

# Функция для обработки сообщений в чате
def process_message(update: Update, context: CallbackContext):
    try:
        # Извлекаем текст сообщения из обновления
        message_text = update.message.text

        # Отправляем действие "typing" для отображения "Печатает"
        bot.send_chat_action(chat_id=update.effective_chat.id, action=ChatAction.TYPING)

        # Ваша логика обработки сообщений здесь
        response = g4f.ChatCompletion.create(
            model="gpt-3.5-turbo",
            messages=[{"role": "user", "content": message_text}],
            stream=True)

        bot_response = ""
        for message in response:
            bot_response += message + ""  # Добавляем сообщение бота

        # Отправляем сообщение в чат
        bot.send_message(chat_id=update.effective_chat.id, text=bot_response)
    except Exception as e: # Блок кода для обработки других исключений (Exception - базовый класс для всех исключений)
        # Отправляем сообщение об ошибке в чат
        bot.send_message(chat_id=update.effective_chat.id, text=f"Сорян,Я не вывожу Ваш запрос: {str(e)}")

# Токен телеграм-бота
TOKEN = '6368420018:AAEA5twfklDSaxspo7XsO6DeSVhShgBKyBg'
bot = telegram.Bot(token=TOKEN)

# Создаем Updater и передаем ему токен вашего бота
updater = Updater(token=TOKEN, use_context=True)

# Получаем диспетчер для регистрации обработчиков
dispatcher = updater.dispatcher

# Регистрируем обработчик сообщений
message_handler = MessageHandler(Filters.text & (~Filters.command), process_message)
dispatcher.add_handler(message_handler)

# Запускаем бота
updater.start_polling()
updater.idle()


#pip install -U g4f
#pip install python-telegram-bot==13
#For windows: Go to %appdata% AppData\Local\Programs\Python\Python311\Lib\site-packages\g4f\Provider Open init.py and remove from .NoowAi import NoowAi line
```

```python
import requests
from telegram.ext import Updater, CommandHandler, MessageHandler, Filters, CallbackContext
from telegram import Update

# Замените 'YOUR_BOT_TOKEN' на ваш токен бота
TOKEN = '6737467049:AAFC80kAD-MWFVpJhi-kkvSnIRggnxl-Lg4'

# Замените 'YOUR_CHATGPT_API_KEY' на ваш API-ключ для ChatGPT
CHATGPT_API_KEY = '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'
CHATGPT_ENDPOINT = 'https://api.openai.com/v1/chat/completions'

def start(update: Update, context: CallbackContext):
    update.message.reply_text("Привет! Я бот на основе GPT. Отправь мне текст, и я постараюсь ответить!")

def generate_gpt_response(prompt):
    headers = {
        'Content-Type': 'application/json',
        'Authorization': f'Bearer {CHATGPT_API_KEY}',
    }
    data = {
        'model': 'gpt-3.5-turbo',
        'messages': [{'role': 'system', 'content': 'You are a helpful assistant.'}, {'role': 'user', 'content': prompt}],
    }
    
    try:
        response = requests.post(CHATGPT_ENDPOINT, headers=headers, json=data)
        response.raise_for_status()  # Raises an HTTPError if the HTTP request returned an unsuccessful status code
        
        # Check if the response contains JSON data
        content_type = response.headers.get('content-type')
        if 'application/json' in content_type.lower():
            return response.json()['choices'][0]['message']['content']
        else:
            return f"Unexpected response content type: {content_type}"
    
    except requests.exceptions.HTTPError as errh:
        return f"HTTP Error: {errh}"
    except requests.exceptions.ConnectionError as errc:
        return f"Error Connecting: {errc}"
    except requests.exceptions.Timeout as errt:
        return f"Timeout Error: {errt}"
    except requests.exceptions.RequestException as err:
        return f"Request Exception: {err}"


def echo(update: Update, context: CallbackContext):
    user_input = update.message.text
    response = generate_gpt_response(user_input)
    update.message.reply_text(response)

def main():
    updater = Updater(TOKEN, use_context=True)
    dp = updater.dispatcher

    dp.add_handler(CommandHandler("start", start))
    dp.add_handler(MessageHandler(Filters.text & ~Filters.command, echo))

    updater.start_polling()
    updater.idle()

if __name__ == '__main__':
    main()
```
