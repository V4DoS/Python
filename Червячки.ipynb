```python
import pygame        #Плюс появление новых квадратов
import sys
import random

# Инициализация Pygame
pygame.init()

# Размеры окна
WIDTH, HEIGHT = 800, 600
# Цвета
GRAY = (200, 200, 200)
GREEN = (0, 255, 0)
RED = (255, 0, 0)
YELLOW = (255, 255, 0)

# Создание окна
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Эволюция")

# Функция отрисовки квадратика
def draw_square(x, y, color):
    pygame.draw.rect(screen, color, (x, y, 7, 7))

# Основной игровой цикл
def main():
    # Начальные координаты квадратиков
    x_green, y_green = WIDTH // 2, HEIGHT // 2
    x_red, y_red = WIDTH // 2, HEIGHT // 2 + 20  # Начальное положение красного квадрата
    # Список координат всех квадратиков
    green_squares = [(x_green, y_green)]
    red_squares = [(x_red, y_red)]
    yellow_squares = []  # Начально нет желтых квадратов

    # Счетчики шагов и квадратов
    step_count = 0
    green_square_count = 1
    red_square_count = 1
    yellow_square_count = 0

    while True:
        # Обработка событий
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit()
                sys.exit()

        # Очистка экрана
        screen.fill(GRAY)

        # Рисуем все квадратики из списка зеленых
        for square in green_squares:
            draw_square(square[0], square[1], GREEN)
        
        # Рисуем все квадратики из списка красных
        for square in red_squares:
            draw_square(square[0], square[1], RED)

        # Рисуем все квадратики из списка желтых
        for square in yellow_squares:
            draw_square(square[0], square[1], YELLOW)

        # Генерация новых координат для следующего квадратика зеленого цвета
        dx, dy = random.choice([(0, 10), (0, -10), (10, 0), (-10, 0)])
        new_x, new_y = green_squares[-1][0] + dx, green_squares[-1][1] + dy
        # Проверка, чтобы квадраты не выходили за пределы экрана
        if 0 <= new_x < WIDTH and 0 <= new_y < HEIGHT:
            green_squares.append((new_x, new_y))
            step_count += 1
            green_square_count += 1

        # Генерация новых координат для следующего квадрата красного цвета
        dx, dy = random.choice([(0, 10), (0, -10), (10, 0), (-10, 0)])
        new_x, new_y = red_squares[-1][0] + dx, red_squares[-1][1] + dy
        # Проверка, чтобы квадраты не выходили за пределы экрана
        if 0 <= new_x < WIDTH and 0 <= new_y < HEIGHT:
            red_squares.append((new_x, new_y))
            step_count += 1
            red_square_count += 1

        # Проверка пересечения красного и зеленого квадрата
        for green_square in green_squares:
            for red_square in red_squares:
                if green_square == red_square:  # Если квадраты пересекаются
                    x, y = green_square[0], green_square[1]
                    yellow_squares.append((x, y))  # Добавляем новый желтый квадрат
                    yellow_square_count += 1

        # Ограничение количества квадратиков для экономии памяти
        if len(green_squares) > 200:
            green_squares.pop(0)
        if len(red_squares) > 200:
            red_squares.pop(0)
        if len(yellow_squares) > 1000:
            yellow_squares.pop(0)


        #Отображение счетчика шагов и квадратов
        font = pygame.font.Font(None, 36)
        step_text = font.render("Шаги: " + str(step_count), True, (0, 0, 0))
#         green_square_text = font.render("Зеленые квадраты: " + str(green_square_count), True, (0, 0, 0))
#         red_square_text = font.render("Красные квадраты: " + str(red_square_count), True, (0, 0, 0))
#         yellow_square_text = font.render("Желтые квадраты: " + str(yellow_square_count), True, (0, 0, 0))
        screen.blit(step_text, (10, 10))
#         screen.blit(green_square_text, (10, 50))
#         screen.blit(red_square_text, (10, 90))
#         screen.blit(yellow_square_text, (10, 130))

        # Обновление экрана
        pygame.display.flip()
        # Задержка для наглядности
        pygame.time.delay(100)

if __name__ == "__main__":
    main()
```
