```python
%%time
import warnings
import pandas as pd             
from prophet import Prophet

warnings.filterwarnings("ignore")

# Блок 1: Загрузка данных и предварительная обработка
data = pd.read_csv('Го data.csv', sep=';', low_memory=False)
data['Дата'] = pd.to_datetime(data['Дата'], dayfirst=True)

# Подготовка данных для модели Prophet
data = data.rename(columns={'Дата': 'ds', 'Магазин': 'shop', 'Склад1': 'Store1', 'Склад2': 'Store2', 'Отгрузка': 'y'})
data['y'] = data['y'].str.replace(',', '.').astype(float)

# Блок 2: Создание и обучение моделей Prophet для каждого магазина и склада
forecast_dfs = []
for shop, store1, store2 in data[['shop', 'Store1', 'Store2']].drop_duplicates().itertuples(index=False):
    # Выбор данных для конкретного магазина и склада
    subset = data[(data['shop'] == shop) & (data['Store1'] == store1) & (data['Store2'] == store2)]
    
    # Проверка на наличие данных для обучения модели
    if len(subset) >= 2:
        # Создание и обучение модели Prophet
        model = Prophet()
        model.fit(subset)
        
        # Создание фрейма данных для прогноза
        future = model.make_future_dataframe(periods=40)  # Прогноз на 30 дней вперед
        
        # Получение прогноза
        forecast = model.predict(future)
        
        # Добавление прогноза в список
        forecast['shop'] = shop
        forecast['Store1'] = store1
        forecast['Store2'] = store2
        forecast_dfs.append(forecast)

# Объединение прогнозов в один DataFrame
final_forecast = pd.concat(forecast_dfs, ignore_index=True)

# Блок 3: Обработка отрицательных прогнозов
final_forecast['yhat'] = final_forecast['yhat'].apply(lambda x: 0 if x < 0 else x)
final_forecast['yhat_upper'] = final_forecast['yhat_upper'].apply(lambda x: 0 if x < 0 else x)

# Блок 4: Сохранение прогноза в Excel с выбранными полями
final_forecast[['ds', 'shop', 'Store1', 'Store2', 'yhat', 'yhat_upper']].to_excel('прогноз.xlsx', index=False)
```
