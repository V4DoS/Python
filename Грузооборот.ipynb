```python
%%time
import warnings
import pandas as pd             
from prophet import Prophet

warnings.filterwarnings("ignore")

#  1:     
data = pd.read_csv(' data.csv', sep=';', low_memory=False)
data[''] = pd.to_datetime(data[''], dayfirst=True)

#     Prophet
data = data.rename(columns={'': 'ds', '': 'shop', '1': 'Store1', '2': 'Store2', '': 'y'})
data['y'] = data['y'].str.replace(',', '.').astype(float)

#  2:     Prophet     
forecast_dfs = []
for shop, store1, store2 in data[['shop', 'Store1', 'Store2']].drop_duplicates().itertuples(index=False):
    #       
    subset = data[(data['shop'] == shop) & (data['Store1'] == store1) & (data['Store2'] == store2)]
    
    #       
    if len(subset) >= 2:
        #     Prophet
        model = Prophet()
        model.fit(subset)
        
        #     
        future = model.make_future_dataframe(periods=40)  #   30  
        
        #  
        forecast = model.predict(future)
        
        #    
        forecast['shop'] = shop
        forecast['Store1'] = store1
        forecast['Store2'] = store2
        forecast_dfs.append(forecast)

#     DataFrame
final_forecast = pd.concat(forecast_dfs, ignore_index=True)

#  3:   
final_forecast['yhat'] = final_forecast['yhat'].apply(lambda x: 0 if x < 0 else x)
final_forecast['yhat_upper'] = final_forecast['yhat_upper'].apply(lambda x: 0 if x < 0 else x)

#  4:    Excel   
final_forecast[['ds', 'shop', 'Store1', 'Store2', 'yhat', 'yhat_upper']].to_excel('.xlsx', index=False)
```
