```python
import random          #Сетка 8 на 8
import pygame
import sys

# Определение констант
GRID_SIZE = 16
CELL_SIZE = 50
WINDOW_WIDTH = GRID_SIZE * CELL_SIZE
WINDOW_HEIGHT = GRID_SIZE * CELL_SIZE
BLUE = (0, 0, 255)
RED = (255, 0, 0)
BLACK = (0, 0, 0)
WHITE = (255, 255, 255)
GREEN = (0, 255, 0)

# Класс для представления бота
class Bot:
    def __init__(self, x, y):
        self.x = x
        self.y = y
        self.energy = 1000
        self.color = BLUE

    def move(self):
        if self.energy > 0:
            self.energy -= 1
            direction = random.choice(['up', 'down', 'left', 'right'])
            if direction == 'up' and self.y > 0:
                self.y -= 1
            elif direction == 'down' and self.y < GRID_SIZE - 1:
                self.y += 1
            elif direction == 'left' and self.x > 0:
                self.x -= 1
            elif direction == 'right' and self.x < GRID_SIZE - 1:
                self.x += 1

    def eat(self):
        if (self.x, self.y) == food_location:
            self.energy += 100

    def check_collision(self):
        if (self.x, self.y) == poison_location:
            if random.random() < 0.7:
                self.color = BLACK
            else:
                self.energy = 0

    def draw(self, surface):
        pygame.draw.rect(surface, self.color, (self.x * CELL_SIZE, self.y * CELL_SIZE, CELL_SIZE, CELL_SIZE))
        font = pygame.font.Font(None, 24)
        text = font.render(str(self.energy), True, BLACK)
        surface.blit(text, (self.x * CELL_SIZE + 5, self.y * CELL_SIZE + 5))

# Создание случайной позиции для еды
def create_food():
    return [(random.randint(0, GRID_SIZE - 1), random.randint(0, GRID_SIZE - 1)) for _ in range(10)]

# Создание случайной позиции для яда
def create_poison():
    return random.randint(0, GRID_SIZE - 1), random.randint(0, GRID_SIZE - 1)

# Инициализация pygame
pygame.init()
window = pygame.display.set_mode((WINDOW_WIDTH, WINDOW_HEIGHT))
pygame.display.set_caption('Эволюция ботов')

# Создание ботов
bots = [Bot(random.randint(0, GRID_SIZE - 1), random.randint(0, GRID_SIZE - 1)) for _ in range(3)]

# Создание еды и яда
food_locations = create_food()
poison_location = create_poison()

# Основной игровой цикл
while True:
    window.fill(WHITE)

    # Рисование сетки
    for x in range(GRID_SIZE):
        pygame.draw.line(window, BLACK, (x * CELL_SIZE, 0), (x * CELL_SIZE, WINDOW_HEIGHT))
        pygame.draw.line(window, BLACK, (0, x * CELL_SIZE), (WINDOW_WIDTH, x * CELL_SIZE))

    # Рисование еды и яда
    for food_location in food_locations:
        pygame.draw.rect(window, GREEN, (food_location[0] * CELL_SIZE, food_location[1] * CELL_SIZE, CELL_SIZE, CELL_SIZE))
    pygame.draw.rect(window, RED, (poison_location[0] * CELL_SIZE, poison_location[1] * CELL_SIZE, CELL_SIZE, CELL_SIZE))

    # Движение и отрисовка ботов
    for bot in bots:
        bot.move()
        bot.eat()
        bot.check_collision()
        bot.draw(window)

    # Обработка событий
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            sys.exit()

    pygame.display.update()
    pygame.time.Clock().tick(10)  # Устанавливаем FPS 10
```
